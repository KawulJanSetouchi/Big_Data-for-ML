````python
# 抽出処理の実行(本来の要件)
for f in files:
    df = pd.read_csv(f)
    # 滞在は除く
    nstay_df = df[df['ODRouteType'] != 'K']
    # メッシュコードが同じものを除く
    nsamemesh_df = nstay_df[nstay_df['start_mesh_area'] != nstay_df['end_mesh_area']]
    # 推定人数が秘匿のところは空白に置換える
    nsamemesh_df.loc[nsamemesh_df['population'] == '***', 'population'] = ''
    # 抽出したものを保存
    nsamemesh_df.to_csv('./LocationDataEx/' + os.path.split(f)[1])
    # 推定人数のユニーク数についても保存
    nsamemesh_df['population'].value_counts().to_csv(('./LocationDataEx/pop_' + os.path.split(f)[1]))
    # 変数の削除
    del df, nstay_df, nsamemesh_df
    gc.collect()
````

````python
files = glob.glob("./LocationDataDistinct_EX/*.csv")
````

````python
# マージ処理（12ファイル = 3h）
cnt = 0
names = ['a.csv', 'b.csv', 'c.csv', 'd.csv', 'e.csv', 'f.csv', 'g.csv', 'h.csv']
itr_names = iter(names)
file_names = []
con_df = None
for f in files:
    # ファイル名の一時保存（12ファイル分）
    file_names.append(f)
    if cnt == 0:
        base_df = pd.read_csv(f, index_col=0)
    elif cnt == 1:
        df = pd.read_csv(f, index_col=0)
        con_df = pd.concat([base_df, df], ignore_index=True)
        con_df.drop_duplicates(inplace=True) # 重複行の削除
    else:
        df = pd.read_csv(f, index_col=0)
        con_df = pd.concat([con_df, df], ignore_index=True)
        con_df.drop_duplicates(inplace=True) # 重複行の削除
    
    if cnt == 11:
        name = next(itr_names)
        # 保存処理
        con_df.to_csv('./221205_Test_Middle/' + name)
        # 出力
        pd.Series(file_names).to_csv('./221205_Test_Middle/fnames_' + name)
        # カウントクリア
        cnt = 0
        # ファイル名のクリア
        file_names.clear()
    else:
        cnt += 1
````
